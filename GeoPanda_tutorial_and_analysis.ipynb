 # Using GeoSeries

```python
from shapely.geometry import Point, Polygon
import geopandas as gpd

# Create a Geoseries of Points
Points = gpd.GeoSeries([Point(1,1), Point(2,2), Point(3,3)])

#Visualize the points
Points.plot()
```

# create a GeoDataFrame from a list of Shapely geometries

```python
#Define some points and polygos
polygon = Polygon ([(0,0), (1,1), (1,0)])
points = [Point(0.5, 0.5), Point (1.5, 1.5)]

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame({'geometry': points + [polygon]})

# plot the GeoDataFrame
gdf.plot(alpha=0.5, linewidth=2, edgecolor='k', color='cyan')
```

# Filtering Data

```python
# Filter the GeoDataFrame for countries in North America
north_america = world[world['continent'] =='North America']

# Plot the filtered GeoDataFrame
north_america.plot()
plt.show()
```

# Basic Spatial Operations

```python
# Convert geometries from geographic coordinate reference system (CRS) 
# In this case we Convert to EPSG:3395: WGS 84 / World Mercator

north_america = north_america.to_crs("EPSG:3395")

#If you want to see if it has changed, you can print it out.
print(north_america.crs)
```

```python
# View the header of the table to see how the columns are saved for the country name and Area in North America

print(north_america.head())
```

```python
# Calculate the area of each country
# Print the area of each country

north_america['area'] = north_america.geometry.area

for index, row in north_america.iterrows():
    print(f"North America: {row['name']}, Area: {row['area']}")
```

```python
# This demonstrates a buffer operation, which creates a buffer area around the geometries in the GeoDataFrame.

buffered_point = north_america.geometry.buffer(1)

print(buffered_point)
```

# Spatial Joins

```python
# Create a GeoDataFrame of cities
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))

# Perform a spatial join between cities and countries
cities_with_country = gpd.sjoin(cities, world, predicate='within')

# Inspect the result
cities_with_country.head()
```

# Overlay Analysis

```python
import geopandas as gpd

# Load two GeoDataFrames representing different layers
rivers = gpd.read_file(r"C:\Users\Owner\Desktop\lab1_data\Rivers.shp")
UScounties = gpd.read_file(r"C:\Users\Owner\Desktop\lab1_data\UScounties.shp")


# Perform an overlay analysis to find intersection
intersection = gpd.overlay(rivers, UScounties, how='intersection')

# Plot the result
intersection.plot()
```

# Buffer Analysis

```python
# Before using the buffer, Convert geometries from geographic coordinate reference system (CRS) 
# In this case we Convert to EPSG:3395: WGS 84 / World Mercator
rivers = rivers.to_crs("EPSG:3395")

# Create a buffer of 1 km around each point in a GeoDataFrame
buffered_points = rivers.geometry.buffer(1000)

# Plot the buffered geometries
buffered_points.plot()
```
